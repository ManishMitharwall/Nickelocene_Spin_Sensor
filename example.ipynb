# Nickelocene as Spin Sensor

In this example, it is shown that how this code can be used in jupyter notebook notebook.

Parameters are defined and later on we can call the kernel() to run the code. 

```python
from Spin_Sensing_Nc import *
```

```python
my = Nc_sensing()
my.M_spin = 3/2           # Spin of Metal        
my.D_aniso = [0,0,5]      # Magneitc anisotropy of Metal [D_x, D_y, D_z]
my.Temp = 0.5             # Temperature
my.B_field = 0.0          # Magnetic Field
my.J_Nc = 3               # Magnetic Exchange between metal and Nc
my.Nsites = 1             # Number of spin sites
my.contrast = 0.7         # Contrast for ploting and should be between 0 and 1 ideal value is between 0.7 to 0.9
my.J1 = 6                 # Nearest Neighbour exchange when Nsites are greater than 1
my.J2 = 0                 # second Neighbour exchange when Nsites are greater than 2
my.States_DOS = 3         # Number of energy states consider in ploting
my.energy = 15            # Energy range for ploting
my.coff_Nc = 3            # coupling of nickelocene spin to the metallic tip appex
my.coff_M  = 1            # coupling of surface spin to the underlying substrate
my.ciclo = 0              # For many spin site if you want to connect last site to first site put 1.


my.kernel()
```
